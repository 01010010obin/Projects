```python
import pandas as pd
from sklearn.model_selection import train_test_split

```

```python

# Örnek veri seti
train_df = pd.read_csv("train.csv")  # Veri setinizi yükleyin

train_df.head()
```

```python
onehot_columns=["SubscriptionType","PaymentMethod","PaperlessBilling","ContentType","MultiDeviceAccess","DeviceRegistered","GenrePreference","Gender","ParentalControl","SubtitlesEnabled"]
```

```python
onehot_df=pd.get_dummies(train_df[onehot_columns],dtype=int)
onehot_df.head()
```

```python
newtrain_df=pd.concat([train_df,onehot_df],axis=1)
newtrain_df.head()
```

```python
newtrain_df.drop(onehot_columns,axis=1,inplace=True)
newtrain_df.head()
```

```python
newtrain_df.set_index("CustomerID",inplace=True)
newtrain_df.index.name=None
```

```python
scaled_columns=['AccountAge', 'MonthlyCharges', 'TotalCharges', 'ViewingHoursPerWeek',
       'AverageViewingDuration', 'ContentDownloadsPerMonth', 'UserRating',
       'SupportTicketsPerMonth', 'WatchlistSize']
```

```python
from sklearn.preprocessing import MinMaxScaler

# Min-Max Scaler 
scaler = MinMaxScaler()

df_scaled = scaler.fit_transform(newtrain_df[scaled_columns])

df_scaled = pd.DataFrame(df_scaled, columns=scaled_columns)
df_scaled.head()
```

```python
droped_newtrain_df=newtrain_df.drop(scaled_columns,axis=1)
droped_newtrain_df.head(1)
```

```python
droped_newtrain_df.reset_index(inplace=True)
droped_newtrain_df=droped_newtrain_df.iloc[:,1:]
droped_newtrain_df.head()
```

```python
con_df=pd.concat([df_scaled,droped_newtrain_df],axis=1)
con_df.head()
```

```python
con_df.info()
```

# Standart Scaler DF

```python
import pandas  as pd
```

```python
stdscaler_df=pd.read_csv("stdScaled_trainDF.csv")
stdscaler_df.head()
```

```python
stdscaler_df=stdscaler_df.iloc[:,1:]
stdscaler_df.head(5)
```

# ---------------------------------------------------------------------------------------------------------------

```python
from sklearn.model_selection import train_test_split
```

```python
# Veriyi eğitim ve test setlerine ayırın
X_train, X_test, y_train, y_test = train_test_split(stdscaler_df.drop(["Churn"],axis=1), stdscaler_df.Churn, test_size=0.2, random_state=42)
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
```

```python
X_train.head(5)
```

```python


# SMOTE nesnesi oluşturun
smote = SMOTE(random_state=42)

# SMOTE ile aşırı örnekleme uygulayın
X_train_res, y_train_res = smote.fit_resample(X_train, y_train)

# Orijinal ve dengelenmiş sınıf dağılımlarını karşılaştırın
print("Orijinal eğitim seti sınıf dağılımı:", Counter(y_train))
print("Dengelenmiş eğitim seti sınıf dağılımı:", Counter(y_train_res))
```

```python


# Dengelenmiş veri setini DataFrame'e dönüştürün
X_train_res_df = pd.DataFrame(X_train_res, columns=X_train.columns)
y_train_res_df = pd.DataFrame(y_train_res, columns=['Churn'])  # Hedef sütununuzun adını 'Churn' olarak değiştirin

# Özellikler ve hedef sütununu birleştirin
balanced_df = pd.concat([X_train_res_df, y_train_res_df], axis=1)
balanced_df.head()
```

```python
balanced_df.shape
```

```python
# Dengelenmiş veri setini CSV dosyasına kaydedin
balanced_df.to_csv('std_balanced_dataset.csv', index=False)
```
